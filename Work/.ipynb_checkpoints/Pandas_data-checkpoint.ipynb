```python
import pandas as pd
```

```python
# Exercise 1
df = pd.DataFrame({
    "order_id": [1, 2, 3, 4, 5],
    "order_date": ["2023-01-01", "2023-01-05", "2023-01-07", "2023-01-15", "2023-01-20"],
    "amount": [100, 150, 200, 130, 90]
})
```

```python
# Convertis la colonne order_date en format datetime

df["order_date"] = pd.to_datetime(df["order_date"])

print("Types des colonnes:")
print(df.dtypes)

print("\nDataFrame avec dates converties:")

df
```

```python
# Ajoute une colonne weekday avec le jour de la semaine

df["weekday"] = df["order_date"].dt.day_name()

df
```

```python
# Calcule le total des ventes par jour de la semaine

df.groupby("weekday")["amount"].sum()

print("\nTotal des ventes par jour de la semaine :")
df
```

```python
#Exercise 2
clients = pd.DataFrame({
    "client_id": [1, 2, 3],
    "nom": ["Alice", "Bob", "Clara"]
})

commandes = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "client_id": [1, 1, 2, 3],
    "montant": [50, 75, 100, 30]
})
```

```python
# Effectue une jointure (merge) entre les clients et leurs commandes

df_merged = pd.merge(clients, commandes)
print("Jointure entre clients et commandes :")
df_merged
```

```python
# Calcule le total des dépenses par client

total_depenses = df_merged.groupby("nom")["montant"].sum().reset_index()

print("\nTotal des dépenses par client :")
total_depenses
```

```python
# Trie les clients par dépense totale décroissante

total_depenses_sorted = total_depenses.sort_values(by = "montant", ascending=False)

print("\nTotal des dépenses par client (tri décroissant) :")
total_depenses_sorted
```

```python
#Exercise 3

df = pd.DataFrame({
    "date": pd.date_range("2023-01-01", periods=6),
    "temperature": [10, None, None, 15, None, 20]
})
```

```python
# Remplis les NaN par la moyenne des températures connues

moyenne = float( df["temperature"].mean() ) #calcul la moyenne avec les valeurs qui sont pas NaN
df.fillna({"temperature" : moyenne}, inplace = True)

df
#Remplis les NaN par interpolation linéaire.
#Quelle différence entre les deux résultats ? 
```

```python
 # Remplis les NaN par interpolation linéaire.

df = pd.DataFrame({
    "date": pd.date_range("2023-01-01", periods=6),
    "temperature": [10, None, None, 15, None, 20]
})

df["temperature"] = df["temperature"].interpolate(method="linear")  # Interpolation linéaire

df

#Quelle différence entre les deux résultats ? 
```

```python
#Exercise 4

df = pd.DataFrame({
    "année": [2022, 2022, 2023, 2023],
    "ville": ["Paris", "Lyon", "Paris", "Lyon"],
    "ventes": [200, 150, 300, 100]
})
```

```python
#Crée un tableau croisé (pivot) avec les années en ligne, villes en colonnes, et la somme des ventes.#Calcule l’évolution en pourcentage des ventes entre 2022 et 2023 pour chaque ville.

pivot = pd.pivot_table(df, index="année", columns="ville", values="ventes", aggfunc="sum")
pivot
```

```python
#Calcule l’évolution en pourcentage des ventes entre 2022 et 2023 pour chaque ville.

t_2022 = float(pivot.loc[2022].sum())
t_2023 = float(pivot.loc[2023].sum())

percentage = ((t_2023 - t_2022) / t_2022) * 100
print(f"{percentage : 0.2f} %")
```

```python
#Exercise 5

df = pd.DataFrame({
    "produit": ["Livre", "Stylo", "Ordinateur", "Clé USB"],
    "prix_ht": [20, 1.5, 900, 10],
    "TVA": [0.05, 0.2, 0.2, 0.2]
})
```

```python
# Crée une colonne prix_ttc en appliquant une fonction via apply

df["prix_ttc"] = df.apply(lambda row: row["prix_ht"] * (1 + row["TVA"]), axis=1)

df

#def squre(x):
    #return x**2
    
#df["prix_ttc"] = df["prix_ht"] * (1 + df["TVA"])

#df

#Applique-la pour créer une colonne catégorie
```

```python
# Crée une fonction qui classe les produits en "bon marché", "moyen" ou "cher" selon le prix TTC

def class_produit(prix_ttc):
    if prix_ttc < 20: 
        return "bon marché"
    elif prix_ttc <= 100:
        return "moyenne"
    else: 
        return "cher"
```

```python
df["catégorie"] = df["prix_ttc"].apply(class_produit)

df
```
